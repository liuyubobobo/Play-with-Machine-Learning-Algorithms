## Pandas 中的数据索引和选择

### Series 中的数据选择

#### 字典式的数据选择

```python
import pandas as pd
data = pd.Series([0.25, 0.5, 0.75, 1.0], 
                 index=['a', 'b', 'c', 'd'])
data
```

```python
data['b']
```

```python
'a' in data
```

```python
data.keys()
```

```python
data.items()
```

```python
list(data.items())
```

像字典操作一样在Series中添加数据

```python
data['e'] = 1.25
```

```python
data
```

#### 一维数组式的数据选择

```python
data['a': 'c'] # 注意，使用显式的index，前后索引都包含在结果中
```

```python
data[1: 3] # 注意，使用引式的数字索引，结果是前闭后开的形式
```

```python
data[(data > 0.3) & (data < 0.8)]
```

```python
data[['a', 'e']]
```

#### ``loc`` 和 ``iloc``

显式索引和隐式索引“冲突”的问题

```python
data = pd.Series(['a', 'b', 'c'], index=[1, 3, 5])
data
```

```python
data[1] # 取单独元素，使用显式索引
```

```python
data[1:3] # 取切片，使用隐式索引
```

**``loc``永远使用显式索引**

```python
data.loc[1]
```

```python
data.loc[1:3]
```

**``iloc``永远使用显式索引**

```python
data.iloc[1]
```

```python
data.iloc[1:3]
```

通常认为：显式比隐式好

### DataFrame 中的数据选择

#### 字典式的数据选择

```python
area = pd.Series({'California': 423967, 'Texas': 695662, 
                  'New York': 141297, 'Florida': 170312,
                  'Illinois': 149995})
pop = pd.Series({'California': 38332521, 'Texas': 26448193, 
                 'New York': 19651127, 'Florida': 19552860,
                 'Illinois': 12882135})
data = pd.DataFrame({'area':area, 'pop':pop})
data
```

```python
data['area']
```

```python
data.area
```

```python
data.area is data['area']
```

```python
data.pop is data['pop'] # 因为pop是一个方法名
```

```python
type(data.pop)
```

使用字典的方式为 DataFrame 添加一列数据

```python
data['density'] = data['pop'] / data['area']
```

```python
data
```

#### 二维数组式的数据选择

```python
data.values
```

```python
data.T
```

```python
data.values[0] # 选择一行
```

```python
data['area'] # 选择一列
```

使用 **``iloc``** 和 **``loc``**

```python
data.iloc[:3, :2]
```

```python
data.loc[:'Illinois', :'pop']
```

```python
data.loc[data.density > 100, ['pop', 'density']]
```

```python
data.iloc[0, 2] = 90
data
```

#### 注意

切片的选择方式都是选择行

```python
data['Florida': 'Illinois']
```

```python
data[1:3]
```

不过为了清晰起见，还是建议使用 **``loc``** 或者 **``iloc``**

```python
data.loc['Florida': 'Illinois']
```

```python
data.iloc[1:3]
```

在DataFrame中使用masking的方式，也是基于行做选择的

```python
data[data.density > 100]
```
